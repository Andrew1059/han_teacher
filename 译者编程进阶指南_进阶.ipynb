````python
# --- 定义函数部分 ---
def translate_en_to_zh(text):
    """英文 → 中文"""
    response = client.chat.completions.create(
        model="glm-4",
        messages=[
            {"role": "user", "content": f"请将下面的英文句子翻译成中文：{text}"}
        ],
    )
    return response.choices[0].message.content


def translate_zh_to_en(text):
    """中文 → 英文"""
    response = client.chat.completions.create(
        model="glm-4",
        messages=[
            {"role": "user", "content": f"请将下面的中文句子翻译成英文：{text}"}
        ],
    )
    return response.choices[0].message.content


# --- 数据部分 ---
english_quotes = [
    "To be, or not to be: that is the question. (Hamlet)",
    "All the world's a stage, and all the men and women merely players. (As You Like It)",
    "What's in a name? That which we call a rose by any other name would smell as sweet. (Romeo and Juliet)",
    "The fault, dear Brutus, is not in our stars, but in ourselves. (Julius Caesar)",
    "Brevity is the soul of wit. (Hamlet)",
    "Some are born great, some achieve greatness, and some have greatness thrust upon them. (Twelfth Night)",
    "Cowards die many times before their deaths. (Julius Caesar)",
    "The better part of Valour, is Discretion. (Henry IV)",
    "Love looks not with the eyes, but with the mind. (A Midsummer Night's Dream)",
    "There is nothing either good or bad, but thinking makes it so. (Hamlet)"
]

chinese_sentences = [
    "生命如戏，每个人都是舞台上的演员。",
    "名字并不重要，玫瑰换个名字依然芬芳。",
    "时间就是金钱。",
    "知识就是力量。",
    "失败乃成功之母。",
    "爱情是盲目的。",
    "世上无难事，只怕有心人。",
    "好书是最好的朋友。",
    "光阴似箭，日月如梭。",
    "真理往往掌握在少数人手中。"
]


# --- 调用函数部分 ---
print("=== 英文句子 → 中文翻译 ===\n")
for quote in english_quotes:
    print(f"原文: {quote}")
    print(f"译文: {translate_en_to_zh(quote)}\n")

print("=== 中文句子 → 英文翻译 ===\n")
for sentence in chinese_sentences:
    print(f"原文: {sentence}")
    print(f"译文: {translate_zh_to_en(sentence)}\n")
````

````python
# 使用智谱AI API 抽取并以JSON格式输出
prompt = '''
文章标题：检索增强生成 (RAG)：为大模型外挂“知识库”

在人工智能飞速发展的今天，大语言模型（LLM）如 GPT-4 和 Claude 虽然展现了惊人的对话能力，但它们存在两个显著的缺陷：一是知识截止日期导致的信息滞后，二是容易产生一本正经胡说八道的“幻觉”（Hallucination）现象。为了解决这些问题，一种名为“检索增强生成”（Retrieval-Augmented Generation，简称 RAG）的技术应运而生。

RAG 的核心理念是将大语言模型的生成能力与外部数据的检索能力相结合。这就好比考试时，学生（大模型）不仅可以凭记忆作答，还可以翻阅一本最新的参考书（外部知识库）。其工作流程主要包含三个步骤：索引、检索和生成。

首先是索引（Indexing）环节。我们需要将大量的私有数据或最新文档通过嵌入模型（Embedding Model）转化为高维向量，并存储在向量数据库（Vector Database）中。这一步是为了让计算机能够理解文本的语义关联。

其次是检索（Retrieval）。当用户提出问题时，系统会将问题也转化为向量，并在数据库中进行相似性搜索（Similarity Search），快速找到与问题最相关的几段背景资料。

最后是生成（Generation）。系统将检索到的相关资料作为“提示词上下文”（Context）喂给大语言模型。大模型结合用户的问题和提供的参考资料，生成准确且基于事实的回答。

请你提取这段话中的中文术语，并翻译成英文，给出中文解释，并输出JSON格式，参考样式为："chinese":"","english":"","explanation":""

'''

response = client.chat.completions.create(
    model="glm-4",  # 填写需要调用的模型名称
    messages=[
        {"role": "user", "content": prompt}
    ],
)
print(response.choices[0].message.content)
````

````python
#将百度翻译的官方代码转换成一个函数

import requests
import random
from hashlib import md5
import json

# Set your own appid/appkey.
appid = ''
appkey = ''

# For list of language codes, please refer to `https://api.fanyi.baidu.com/doc/21`
endpoint = 'http://api.fanyi.baidu.com'
path = '/api/trans/vip/translate'

# Function to generate md5 hash
def make_md5(s, encoding='utf-8'):
    return md5(s.encode(encoding)).hexdigest()

# The main translation function
def baidu_translate(query, from_lang, to_lang):
    url = endpoint + path

    # Generate salt and sign
    salt = random.randint(32768, 65536)
    sign = make_md5(appid + query + str(salt) + appkey)

    # Build request
    headers = {'Content-Type': 'application/x-www-form-urlencoded'}
    payload = {'appid': appid, 'q': query, 'from': from_lang, 'to': to_lang, 'salt': salt, 'sign': sign}

    # Send request
    r = requests.post(url, params=payload, headers=headers)
    result = r.json()

    # Extract and return the translated text
    return result['trans_result'][0]['dst']

# Example usage:
translated_text = baidu_translate('北京语言大学高级翻译学院开设了翻译专业本科。', 'zh', 'en')
print(translated_text)
````

````python
#进一步将GLM的代码也变成一个函数，是代码更为简洁

# 假设client是已经初始化的GLM客户端

def glm_translate(source, termlist, reference_str):
    # 构建prompt，包含术语列表和翻译建议
    prompt = f"请将这句话翻译成英文:【{source}】"

    if termlist:
        prompt += f"，并严格参照以下术语:{termlist}"
    if reference_str:
        prompt += f"，并参考以下翻译建议:{reference_str}"
    prompt += "。最后仅输出译文即可。"

    # 调用GLM模型
    response = client.chat.completions.create(
        model="glm-4",  # 填写需要调用的模型名称
        messages=[
            {"role": "user", "content": prompt}
        ],
    )

    # 解析响应并提取翻译结果
    translation = response.choices[0].message.content

    return translation
````
